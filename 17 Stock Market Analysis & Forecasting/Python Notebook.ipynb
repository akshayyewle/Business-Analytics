```python
import os 
import pandas as pd 
import numpy as np 
import yfinance as yf
from pandas_datareader import data as pdr
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use('fivethirtyeight')
```

# **Data Import**

```python
Stock_Ticker_List = ['AAPL','GOOG','MSFT','AMZN','META']

# Today's Time
Today_Time = datetime.now()
print('Today\'s Time',Today_Time)

# Start Time
# Start_Time = datetime((Today_Time.year - 1),Today_Time.month, Today_Time.day)
Start_Time = Today_Time - timedelta(days=366)
print('Start Time: ',Start_Time)
```

## **Download Data**

```python
for stock in Stock_Ticker_List:
    globals() [stock] = yf.download(stock,start=Start_Time,end=Today_Time)
```

## **Review Data**

```python
# Display Contents Of Data Frame

display(AAPL.head())
display(MSFT.head())
display(GOOG.head())
display(AMZN.head())
display(META.head())
```

## **Modify Data**

* Add Company Name Column & Concatenate Dataframes

```python
Stock_Df_List = [AAPL,MSFT,GOOG,AMZN,META]
Stock_Name_List = ['Apple','Microsoft','Google','Amazon','Meta']

for Stock_Df, Stock_Name in zip(Stock_Df_List,Stock_Name_List):
    Stock_Df['Company_Name'] = Stock_Name
    display(Stock_Df.head())
```

## **Concatenate Dataframes**

```python
df = pd.concat(Stock_Df_List,axis=0)

print(df.shape)
print(df.columns)

display(df.head())
display(df.tail())
```

# **Data Exploration**

## **Closing Price**

```python
plt.figure(figsize=(12,15))
plt.subplots_adjust(top=5,bottom=4)

for i,stock in enumerate(Stock_Df_List,start=1):
    plt.subplot(5,1,i)
    plt.plot(stock['Adj Close'])
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f'Closing Price of {Stock_Name_List[i-1]}',loc='left',pad=10)

plt.tight_layout(pad=2,h_pad=2,w_pad=2)
plt.show()
```
