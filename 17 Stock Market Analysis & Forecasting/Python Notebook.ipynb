```python
import os 
import pandas as pd 
import numpy as np 
import yfinance as yf
from pandas_datareader import data as pdr
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use('fivethirtyeight')
```

# **Data Import**

```python
Stock_Ticker_List = ['AAPL','GOOG','MSFT','AMZN']

# Today's Time
Today_Time = datetime.now()
print('Today\'s Time',Today_Time)

# Start Time
# Start_Time = datetime((Today_Time.year - 1),Today_Time.month, Today_Time.day)
Start_Time = Today_Time - timedelta(days=366)
print('Start Time: ',Start_Time)
```

## **Download Data**

```python
for stock in Stock_Ticker_List:
    globals() [stock] = yf.download(stock,start=Start_Time,end=Today_Time)
```

## **Review Data**

```python
# Display Contents Of Data Frame

display(AAPL.head())
display(MSFT.head())
display(GOOG.head())
display(AMZN.head())
```

## **Modify Data**

* Add Company Name Column & Concatenate Dataframes

```python
Stock_Df_List = [AAPL,MSFT,GOOG,AMZN]
Stock_Name_List = ['Apple','Microsoft','Google','Amazon']

for Stock_Df, Stock_Name in zip(Stock_Df_List,Stock_Name_List):
    Stock_Df['Company_Name'] = Stock_Name
    display(Stock_Df.head())
```

## **Concatenate Dataframes**

```python
df = pd.concat(Stock_Df_List,axis=0)

print(df.shape)
print(df.columns)

display(df.head())
display(df.tail())
```

# **Data Analysis**

## **Closing Price**

```python
plt.figure(figsize=(12,8))
plt.subplots_adjust(top=5,bottom=4)

for i,stock in enumerate(Stock_Df_List,start=1):
    plt.subplot(2,2,i)
    plt.plot(stock['Adj Close'])
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.xticks(rotation=45)
    plt.title(f'{Stock_Name_List[i-1]} - Adj Closing Price',loc='left',pad=10)

plt.tight_layout(pad=2,h_pad=2,w_pad=2)
plt.show()
```

## **Trading Volume**

```python
df.columns
```

```python
plt.figure(figsize=(12,15))
plt.subplots_adjust(top=5,bottom=4)

for i,stock in enumerate(Stock_Df_List,start=1):
    plt.subplot(4,1,i)
    plt.plot(stock['Volume'])
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f'Trading Volume of {Stock_Name_List[i-1]}',loc='left',pad=10)

plt.tight_layout(pad=2,h_pad=2,w_pad=2)
plt.show()
```

## **Moving Avg**

```python
MA_Duration_List = [10,20,30]

for MA_Duration in MA_Duration_List:
    for Stock_Df in Stock_Df_List:
        Stock_Df[f'MA_{MA_Duration}'] = Stock_Df['Adj Close'].rolling(window=MA_Duration).mean()
```

```python
display(AAPL.tail())
```

```python
plt.figure(figsize=(12,25))
plt.subplots_adjust(top=5,bottom=4)

for i,stock in enumerate(Stock_Df_List,start=1):
    plt.subplot(4,1,i)
    plt.plot(stock['MA_10'],label='MA_10')
    plt.plot(stock['MA_20'],label='MA_20')
    plt.plot(stock['MA_30'],label='MA_30')
    plt.ylabel('Avg_Price')
    plt.xlabel(None)
    plt.legend()
    plt.title(f'Moving Avg of {Stock_Name_List[i-1]}',loc='left',pad=10)

plt.tight_layout(pad=2,h_pad=2,w_pad=2)
plt.show()
```

## **Daily Returns - Line Plot**

```python
for Stock_Df in Stock_Df_List:
    Stock_Df['Daily_Returns'] = Stock_Df['Adj Close'].pct_change() * 100
```

```python
display(AAPL.head())
display(MSFT.head())
```

```python
plt.figure(figsize=(12,25))
plt.subplots_adjust(top=5,bottom=4)

for i,stock in enumerate(Stock_Df_List,start=1):
    plt.subplot(4,1,i)
    plt.plot(stock['Daily_Returns'],linestyle='--',marker='o')
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f'Moving Avg of {Stock_Name_List[i-1]}',loc='left',pad=10)

plt.tight_layout(pad=2,h_pad=2,w_pad=2)
plt.show()
```

## **Daily Returns - Histogram**

```python
plt.figure(figsize=(12,10))
plt.subplots_adjust(top=5,bottom=4)

for i,stock in enumerate(Stock_Df_List,start=1):
    plt.subplot(2,2,i)
    plt.hist(stock['Daily_Returns'],bins=100)
    # plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f'{Stock_Name_List[i-1]} - Daily Returns Histogram',loc='left',pad=10)

plt.tight_layout(pad=1,h_pad=1,w_pad=0.9)
plt.show()
```

## **Price Correlation**

```python
print(Stock_Ticker_List)
print(Stock_Name_List)
```

```python
# Stock_AdjClose_Df = pdr.get_data_yahoo(Stock_Ticker_List, start=Start_Time, end= Today_Time)['Close']

Stock_AdjClose_Df = pd.DataFrame()

for Stock_Ticker in Stock_Ticker_List:
    Stock_AdjClose_Df[Stock_Ticker] = yf.download(Stock_Ticker, start=Start_Time, end= Today_Time
                                                  ,progress=False)['Adj Close']

Stock_AdjClose_Df.head(10)
```

```python
Stock_Return_Df = Stock_AdjClose_Df.pct_change() * 100

print(Stock_Return_Df.shape)
Stock_Return_Df.head()
```

## **Jointplot - Google Vs Google**

```python
sns.jointplot(data=Stock_Return_Df,x='GOOG',y='GOOG')
```

## **Joint Plot - GOOG Vs MSFT**

```python
sns.jointplot(data=Stock_Return_Df,x='GOOG',y='MSFT')
```

## **Pairplot - Stock Returns**

```python
sns.pairplot(data=Stock_Return_Df,kind='reg')
plt.show()
```

## **Pairgrid Plot - GOOG Vs MSFT**

```python
# Change Size
# plt.rcParams['figure.figsize'] = [2,2]

PP_Plot01 = sns.PairGrid(data=Stock_Return_Df,height=2, aspect=1,layout_pad=0.05)
PP_Plot01.map_upper(plt.scatter,color='purple')
PP_Plot01.map_lower(sns.kdeplot, cmap="cool_d")
PP_Plot01.map_diag(sns.histplot,bins=50)

plt.tight_layout()
plt.show()
```
